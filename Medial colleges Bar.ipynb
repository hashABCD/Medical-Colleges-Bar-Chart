```python
import pandas as pd
import numpy as np
```

```python
#Scrap data

import requests
from bs4 import  BeautifulSoup

url="http://www.ragingbulledu.com/Edu_Reports/Medicine/Indian_Medical_Colleges_Ranked_by_Primitiveness.aspx"
r=requests.get(url)
soup=BeautifulSoup(r.text, 'html.parser')
rows=soup.find_all('table')[0].find_all('tr')
cols=[x.text.strip() for x in rows[4].find_all('td')]
data=[]
for row in rows[5:]:
    data.append([x.text.strip() for x in row.find_all('td')])
df=pd.DataFrame(data=data, columns=cols)
df=df[df.State!='']
df.to_csv('med_collage.csv', index=None)
```

```python
df=pd.read_csv('med_collage.csv')
df.head()
```

```python
df['Starting Year']=df['Starting Year'].astype('int')
```

```python
df['Starting Year'].min()
```

```python
year=np.arange(df['Starting Year'].min(),df['Starting Year'].max()+1)
```

```python
cols=['State']
#cols.extend(year)
cols
```

```python
temp=pd.DataFrame(columns=cols)
temp
```

```python
temp.State=df.State.unique()
temp.head()
```

```python
def total_counter(year):
    t=pd.DataFrame({'State':list(df[df['Starting Year']<year+1].State.value_counts().index),
                   str(year):list(df[df['Starting Year']<year+1].State.value_counts().values)})
    return t
```

```python
for y in year:
    y_df=total_counter(y)
    temp=temp.merge(y_df, how='outer', on='State')
```
